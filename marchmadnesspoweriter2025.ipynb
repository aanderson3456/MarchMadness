```python
MTeams = pd.read_csv("/Users/austinanderson/GitHub/MarchMadness/MarchMadness/MTeams.csv")
MIDtoName = MTeams.set_index("TeamID")["TeamName"].to_dict()
MNametoID = MTeams.set_index("TeamName")["TeamID"].to_dict()
WTeams = pd.read_csv("/Users/austinanderson/GitHub/MarchMadness/MarchMadness/WTeams.csv")
WIDtoName = WTeams.set_index("TeamID")["TeamName"].to_dict()
WNametoID = WTeams.set_index("TeamName")["TeamID"].to_dict()


MGames = pd.read_csv("/Users/austinanderson/GitHub/MarchMadness/MarchMadness/MRegularSeasonCompactResults.csv")
WGames = pd.read_csv("/Users/austinanderson/GitHub/MarchMadness/MarchMadness/WRegularSeasonCompactResults.csv")

#build matrix for power iteration
Mlen = MTeams.shape[0]
Mmat3 = []
Mmat5 = []
for i in range(Mlen):
    Mmat3.append(list(np.zeros((Mlen,), dtype=int)))
    Mmat5.append(list(np.zeros((Mlen,), dtype=int)))

Wlen = MTeams.shape[0]
#seems to be missing IDs in WTeams
Wmat3 = []
Wmat5 = []
for i in range(Wlen):
    Wmat3.append(list(np.zeros((Wlen,), dtype=int)))
    Wmat5.append(list(np.zeros((Wlen,), dtype=int)))

print("done w cell at" ,datetime.datetime.now().time())
```

```python
M25 = MGames[MGames['Season']==2025]
for index, row in M25.iterrows():
    #print(f"Index: {index}, WTeamID: {row['WTeamID']}, WScore: {row['WScore']}, LTeamID: {row['LTeamID']}, LScore: {row['LScore']}")
    winnerID = row['WTeamID']
    loserID = row['LTeamID']
    winnerLoc = winnerID - 1101
    loserLoc = loserID - 1101
    diff = row['WScore'] - row['LScore']
    Mmat5[winnerLoc][loserLoc] += 5*diff
    Mmat3[winnerLoc][loserLoc] += 3*diff


W25 = WGames[WGames['Season']==2025]
for index, row in W25.iterrows():
    #print(f"Index: {index}, WTeamID: {row['WTeamID']}, WScore: {row['WScore']}, LTeamID: {row['LTeamID']}, LScore: {row['LScore']}")
    winnerID = row['WTeamID']
    loserID = row['LTeamID']
    winnerLoc = winnerID - 3101
    loserLoc = loserID - 3101
    diff = row['WScore'] - row['LScore']
    Wmat5[winnerLoc][loserLoc] += 5*diff
    Wmat3[winnerLoc][loserLoc] += 3*diff


M24 = MGames[MGames['Season']==2024]
for index, row in M24.iterrows():
    #print(f"Index: {index}, WTeamID: {row['WTeamID']}, WScore: {row['WScore']}, LTeamID: {row['LTeamID']}, LScore: {row['LScore']}")
    winnerID = row['WTeamID']
    loserID = row['LTeamID']
    winnerLoc = winnerID - 1101
    loserLoc = loserID - 1101
    diff = row['WScore'] - row['LScore']
    Mmat5[winnerLoc][loserLoc] += 2*diff
    Mmat3[winnerLoc][loserLoc] += 2*diff


W24 = WGames[WGames['Season']==2024]
for index, row in W24.iterrows():
    #print(f"Index: {index}, WTeamID: {row['WTeamID']}, WScore: {row['WScore']}, LTeamID: {row['LTeamID']}, LScore: {row['LScore']}")
    winnerID = row['WTeamID']
    loserID = row['LTeamID']
    winnerLoc = winnerID - 3101
    loserLoc = loserID - 3101
    diff = row['WScore'] - row['LScore']
    Wmat5[winnerLoc][loserLoc] += 2*diff
    Wmat3[winnerLoc][loserLoc] += 2*diff

print("done w cell at" ,datetime.datetime.now().time())
```

```python
#from https://andrewcharlesjones.github.io/journal/power-iteration.html
num_iters = 100
A = np.array(Mmat5)
# Starting vector
b = np.random.rand(A.shape[0])
# Power iteration
for ii in range(num_iters):   
    # Project
    bnew = A @ b    
    # Normalize
    b = bnew / np.linalg.norm(bnew, ord=2)
    
MpowRanks5 = dict()
for i in range(len(b)):
    MpowRanks5[MIDtoName[i+1101]] = b[i]
    #print('team',MIDtoName[i+1101],'value',b[i]*1000)

A = np.array(Wmat5)
# Starting vector
b = np.random.rand(A.shape[0])
# Power iteration
for ii in range(num_iters):   
    # Project
    bnew = A @ b    
    # Normalize
    b = bnew / np.linalg.norm(bnew, ord=2)

WpowRanks5 = dict()
for i in range(len(b)):
    try:
        WpowRanks5[WIDtoName[i+3101]] = b[i]
        #print('team',WIDtoName[i+3101],'value',b[i]*1000)
    except:
        print("looks like",i+3101,"not a WteamID")

#from https://andrewcharlesjones.github.io/journal/power-iteration.html
num_iters = 1000
A = np.array(Mmat3)
# Starting vector
b = np.random.rand(A.shape[0])
# Power iteration
for ii in range(num_iters):   
    # Project
    bnew = A @ b    
    # Normalize
    b = bnew / np.linalg.norm(bnew, ord=2)
    
MpowRanks3 = dict()
for i in range(len(b)):
    MpowRanks3[MIDtoName[i+1101]] = b[i]
    #print('team',MIDtoName[i+1101],'value',b[i]*1000)

A = np.array(Wmat3)
# Starting vector
b = np.random.rand(A.shape[0])
# Power iteration
for ii in range(num_iters):   
    # Project
    bnew = A @ b    
    # Normalize
    b = bnew / np.linalg.norm(bnew, ord=2)

WpowRanks3 = dict()
for i in range(len(b)):
    try:
        WpowRanks3[WIDtoName[i+3101]] = b[i]
        #print('team',WIDtoName[i+3101],'value',b[i]*1000)
    except:
        print("looks like",i+3101,"not a WteamID")
print("done w cell at" ,datetime.datetime.now().time())
```

```python
def simGame(t1name,t2name,gender = 'M',version = 5):
    if gender == "M":
        diff = MpowRanks5[t1name] - MpowRanks5[t2name]
        sumt = MpowRanks5[t1name] + MpowRanks5[t2name]
        prob = 0.5 + 0.5*(diff/sumt)
        #print(t1name,'wins w prob',prob)
    if gender == "W":
        diff = WpowRanks5[t1name] - WpowRanks5[t2name]
        sumt = WpowRanks5[t1name] + WpowRanks5[t2name]
        prob = 0.5 + 0.5*(diff/sumt)
        #print(t1name,'wins w prob',prob)
    if version == 3:
        if gender == "M":
            diff = MpowRanks3[t1name] - MpowRanks3[t2name]
            sumt = MpowRanks3[t1name] + MpowRanks3[t2name]
            prob = 0.5 + 0.5*(diff/sumt)
            #print(t1name,'wins w prob',prob)
        if gender == "W":
            diff = WpowRanks3[t1name] - WpowRanks3[t2name]
            sumt = WpowRanks3[t1name] + WpowRanks3[t2name]
            prob = 0.5 + 0.5*(diff/sumt)
            #print(t1name,'wins w prob',prob)
    return prob
print("done w cell at" ,datetime.datetime.now().time())
```

```python
for i in [3,5]:
    print(simGame("Florida","Houston","M",i))

for i in [3,5]:
    print(simGame("Marquette","New Mexico","M",i))
    
for i in [3,5]:
    print(simGame("St Mary's CA","Vanderbilt","M",i))

    
for i in [3,5]:
    print(simGame("South Carolina","UCLA","W",i))
```
