```python
MTeams = pd.read_csv("/kaggle/input/competitions/march-machine-learning-mania-2026/MTeams.csv")
MIDtoName = MTeams.set_index("TeamID")["TeamName"].to_dict()
MNametoID = MTeams.set_index("TeamName")["TeamID"].to_dict()
WTeams = pd.read_csv("/kaggle/input/competitions/march-machine-learning-mania-2026/WTeams.csv")
WIDtoName = WTeams.set_index("TeamID")["TeamName"].to_dict()
WNametoID = WTeams.set_index("TeamName")["TeamID"].to_dict()


MGames = pd.read_csv("/kaggle/input/competitions/march-machine-learning-mania-2026/MRegularSeasonCompactResults.csv")
WGames = pd.read_csv("/kaggle/input/competitions/march-machine-learning-mania-2026/WRegularSeasonCompactResults.csv")

#build matrix for power iteration
Mlen = MTeams.shape[0]
Mmat3 = []
Mmat5 = []
for i in range(Mlen):
    Mmat3.append(list(np.zeros((Mlen,), dtype=int)))
    Mmat5.append(list(np.zeros((Mlen,), dtype=int)))

Wlen = MTeams.shape[0]
#seems to be missing IDs in WTeams
Wmat3 = []
Wmat5 = []
for i in range(Wlen):
    Wmat3.append(list(np.zeros((Wlen,), dtype=int)))
    Wmat5.append(list(np.zeros((Wlen,), dtype=int)))

print("done w cell at" ,datetime.datetime.now().time())
```

```python
M26 = MGames[MGames['Season']==2026]
for index, row in M26.iterrows():
    #print(f"Index: {index}, WTeamID: {row['WTeamID']}, WScore: {row['WScore']}, LTeamID: {row['LTeamID']}, LScore: {row['LScore']}")
    winnerID = row['WTeamID']
    loserID = row['LTeamID']
    winnerLoc = winnerID - 1101
    loserLoc = loserID - 1101
    diff = row['WScore'] - row['LScore']
    Mmat5[winnerLoc][loserLoc] += 5*diff
    Mmat3[winnerLoc][loserLoc] += 3*diff


W26 = WGames[WGames['Season']==2026]
for index, row in W26.iterrows():
    #print(f"Index: {index}, WTeamID: {row['WTeamID']}, WScore: {row['WScore']}, LTeamID: {row['LTeamID']}, LScore: {row['LScore']}")
    winnerID = row['WTeamID']
    loserID = row['LTeamID']
    winnerLoc = winnerID - 3101
    loserLoc = loserID - 3101
    diff = row['WScore'] - row['LScore']
    Wmat5[winnerLoc][loserLoc] += 5*diff
    Wmat3[winnerLoc][loserLoc] += 3*diff


M25 = MGames[MGames['Season']==2025]
for index, row in M25.iterrows():
    #print(f"Index: {index}, WTeamID: {row['WTeamID']}, WScore: {row['WScore']}, LTeamID: {row['LTeamID']}, LScore: {row['LScore']}")
    winnerID = row['WTeamID']
    loserID = row['LTeamID']
    winnerLoc = winnerID - 1101
    loserLoc = loserID - 1101
    diff = row['WScore'] - row['LScore']
    Mmat5[winnerLoc][loserLoc] += 2*diff
    Mmat3[winnerLoc][loserLoc] += 2*diff


W25 = WGames[WGames['Season']==2025]
for index, row in W25.iterrows():
    #print(f"Index: {index}, WTeamID: {row['WTeamID']}, WScore: {row['WScore']}, LTeamID: {row['LTeamID']}, LScore: {row['LScore']}")
    winnerID = row['WTeamID']
    loserID = row['LTeamID']
    winnerLoc = winnerID - 3101
    loserLoc = loserID - 3101
    diff = row['WScore'] - row['LScore']
    Wmat5[winnerLoc][loserLoc] += 2*diff
    Wmat3[winnerLoc][loserLoc] += 2*diff

print("done w cell at" ,datetime.datetime.now().time())
```

```python
#from https://andrewcharlesjones.github.io/journal/power-iteration.html
num_iters = 100
A = np.array(Mmat5)
# Starting vector
b = np.random.rand(A.shape[0])
# Power iteration
for ii in range(num_iters):   
    # Project
    bnew = A @ b    
    # Normalize
    b = bnew / np.linalg.norm(bnew, ord=2)
    
MpowRanks5 = dict()
for i in range(len(b)):
    MpowRanks5[MIDtoName[i+1101]] = b[i]
    #print('team',MIDtoName[i+1101],'value',b[i]*1000)

A = np.array(Wmat5)
# Starting vector
b = np.random.rand(A.shape[0])
# Power iteration
for ii in range(num_iters):   
    # Project
    bnew = A @ b    
    # Normalize
    b = bnew / np.linalg.norm(bnew, ord=2)

WpowRanks5 = dict()
for i in range(len(b)):
    try:
        WpowRanks5[WIDtoName[i+3101]] = b[i]
        #print('team',WIDtoName[i+3101],'value',b[i]*1000)
    except:
        print("looks like",i+3101,"not a WteamID")

#from https://andrewcharlesjones.github.io/journal/power-iteration.html
num_iters = 1000
A = np.array(Mmat3)
# Starting vector
b = np.random.rand(A.shape[0])
# Power iteration
for ii in range(num_iters):   
    # Project
    bnew = A @ b    
    # Normalize
    b = bnew / np.linalg.norm(bnew, ord=2)
    
MpowRanks3 = dict()
for i in range(len(b)):
    MpowRanks3[MIDtoName[i+1101]] = b[i]
    #print('team',MIDtoName[i+1101],'value',b[i]*1000)

A = np.array(Wmat3)
# Starting vector
b = np.random.rand(A.shape[0])
# Power iteration
for ii in range(num_iters):   
    # Project
    bnew = A @ b    
    # Normalize
    b = bnew / np.linalg.norm(bnew, ord=2)

WpowRanks3 = dict()
for i in range(len(b)):
    try:
        WpowRanks3[WIDtoName[i+3101]] = b[i]
        #print('team',WIDtoName[i+3101],'value',b[i]*1000)
    except:
        print("looks like",i+3101,"not a WteamID")
print("done w cell at" ,datetime.datetime.now().time())
```

```python
from scipy.stats import norm
import datetime

def simGame(t1name, t2name, gender='M', version=5):
    # 1. Select the correct ranking dictionary based on version and gender
    if version == 3:
        ranks = MpowRanks3 if gender == 'M' else WpowRanks3
    else:  # Default to version 5
        ranks = MpowRanks5 if gender == 'M' else WpowRanks5
    
    # 2. Calculate the rating difference
    # If t1 is better, diff is positive
    diff = ranks[t1name] - ranks[t2name]
    
    # 3. Define the 'Scale' (Standard Deviation)
    # In NCAA basketball, a spread of 10-12 points is a standard 'scale'.
    # If your ranks are point-based (e.g., 85.5 vs 80.0), use 11.0.
    # If your ranks are small numbers (e.g., 1.5 vs 1.2), you may need a smaller scale like 0.5.
    scale = 0.5 
    
    # 4. Calculate probability using the Normal CDF
    # norm.cdf(x) returns the area under the bell curve to the left of x
    prob = norm.cdf(diff / scale)
    
    # Optional: Clip the results so you never have 0% or 100% 
    # (Kaggle penalizes heavily for 'confident' wrong answers)
    prob = max(0.01, min(0.99, prob))
    
    return prob

print("done w cell at", datetime.datetime.now().time())
```

```python
import pandas as pd
import datetime

# 1. Load the actual template Kaggle expects
# Change the path if the filename is slightly different (e.g. SampleSubmissionStage1.csv)
sub = pd.read_csv("/kaggle/input/competitions/march-machine-learning-mania-2026/SampleSubmissionStage2.csv")

# 2. Make sure Pred is a float
sub['Pred'] = sub['Pred'].astype(float)

print(f"Starting simulations for {len(sub)} games...")

for index, row in sub.iterrows():
    # ID is "YYYY_ID1_ID2" -> split it into parts
    parts = row['ID'].split('_')
    year = parts[0]
    t1 = int(parts[1])
    t2 = int(parts[2])
    
    # Determine gender and get names
    # Men's IDs are < 3000, Women's are >= 3000
    if t1 < 3000:
        gender = "M"
        t1name = MIDtoName[t1]
        t2name = MIDtoName[t2]
    else:
        gender = "W"
        t1name = WIDtoName[t1]
        t2name = WIDtoName[t2]
    
    # Run your simulation
    prediction = simGame(t1name, t2name, gender)
    
    # Save the prediction to the 'Pred' column
    sub.at[index, 'Pred'] = prediction

# 3. Save the final file
sub.to_csv('submission.csv', index=False)

print("Submission file saved!")
print(sub.head()) # Preview the first few rows
print("done at", datetime.datetime.now().time())
```
